## DISTRIBUCION DE EVASION

### En este paso, el objetivo es comprender cÃ³mo estÃ¡ distribuida la variable "churn" (evasiÃ³n) entre los clientes. Utiliza grÃ¡ficos para visualizar la proporciÃ³n de clientes que permanecieron y los que se dieron de baja.

````python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))

# Colores personalizados y etiquetas con conteo
ax = sns.countplot(x='Churn',  hue='Churn', data=datos_expandido, palette='viridis')

# Agregar los nÃºmeros encima de cada barra
for p in ax.patches:
    ax.annotate(f'{p.get_height():,}',
                (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=12, color='black', xytext=(0, 8), textcoords='offset points')

plt.title('DistribuciÃ³n de Clientes que Abandonan (Churn)', fontsize=16, fontweight='bold')
plt.xlabel('Churn (0 = No, 1 = SÃ­)', fontsize=12)
plt.ylabel('NÃºmero de Clientes', fontsize=12)
sns.despine()

plt.show()
````

## **Recuento de evasiÃ³n por variables categÃ³ricas**

### Ahora, exploraremos cÃ³mo se distribuye la evasiÃ³n segÃºn variables categÃ³ricas, como gÃ©nero, tipo de contrato, mÃ©todo de pago, entre otras

````python
categoricas = ['Genero', 'Contrato', 'Metodo_Pago']

sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

for i, col in enumerate(categoricas, 1):
    plt.subplot(2, 2, i)

    ax = sns.countplot(data=datos_expandido, x=col, hue='Churn',palette="viridis")
    plt.xticks(rotation=45, ha='right')


    for p in ax.patches:
        ax.annotate(f'{p.get_height()}',(p.get_x() + p.get_width() / 2., p.get_height()),ha='center', va='bottom', fontsize=9, color='black')

    plt.title(f'DistribuciÃ³n de {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('NÃºmero de Clientes', fontsize=12)

plt.tight_layout()
plt.show()
````

## **Conteo de evasiÃ³n por variables numÃ©ricas**

### En este paso, explora cÃ³mo las variables numÃ©ricas, como "total gastado" o "tiempo de contrato", se distribuyen entre los clientes que cancelaron (evasiÃ³n) y los que no cancelaron.

````python
variables_numericas = ['Cuentas_Totales', 'Meses_Contrato']

plt.figure(figsize=(12, 5))

for i, var in enumerate(variables_numericas, 1):
    plt.subplot(1, len(variables_numericas), i)
    sns.boxplot(x='Churn', y=var, data=datos_expandido, palette="Set2", hue='Churn')
    plt.title(f'{var} vs Churn')
    plt.xlabel('Churn (0=No, 1=SÃ­)')
    plt.ylabel(var)

plt.tight_layout()
plt.show()
````
