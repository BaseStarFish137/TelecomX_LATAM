## DISTRIBUCION DE EVASION

### En este paso, el objetivo es comprender c√≥mo est√° distribuida la variable "churn" (evasi√≥n) entre los clientes. Utiliza gr√°ficos para visualizar la proporci√≥n de clientes que permanecieron y los que se dieron de baja.

````python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))

# Colores personalizados y etiquetas con conteo
ax = sns.countplot(x='Churn',  hue='Churn', data=datos_expandido, palette='viridis')

# Agregar los n√∫meros encima de cada barra
for p in ax.patches:
    ax.annotate(f'{p.get_height():,}',
                (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=12, color='black', xytext=(0, 8), textcoords='offset points')

plt.title('Distribuci√≥n de Clientes que Abandonan (Churn)', fontsize=16, fontweight='bold')
plt.xlabel('Churn (0 = No, 1 = S√≠)', fontsize=12)
plt.ylabel('N√∫mero de Clientes', fontsize=12)
sns.despine()

plt.show()
````

## **Recuento de evasi√≥n por variables categ√≥ricas**

### Ahora, exploraremos c√≥mo se distribuye la evasi√≥n seg√∫n variables categ√≥ricas, como g√©nero, tipo de contrato, m√©todo de pago, entre otras

````python
categoricas = ['Genero', 'Contrato', 'Metodo_Pago']

sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

for i, col in enumerate(categoricas, 1):
    plt.subplot(2, 2, i)

    ax = sns.countplot(data=datos_expandido, x=col, hue='Churn',palette="viridis")
    plt.xticks(rotation=45, ha='right')


    for p in ax.patches:
        ax.annotate(f'{p.get_height()}',(p.get_x() + p.get_width() / 2., p.get_height()),ha='center', va='bottom', fontsize=9, color='black')

    plt.title(f'Distribuci√≥n de {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('N√∫mero de Clientes', fontsize=12)

plt.tight_layout()
plt.show()
````

## **Conteo de evasi√≥n por variables num√©ricas**

### En este paso, explora c√≥mo las variables num√©ricas, como "total gastado" o "tiempo de contrato", se distribuyen entre los clientes que cancelaron (evasi√≥n) y los que no cancelaron.

````python
variables_numericas = ['Cuentas_Totales', 'Meses_Contrato']

plt.figure(figsize=(12, 5))

for i, var in enumerate(variables_numericas, 1):
    plt.subplot(1, len(variables_numericas), i)
    sns.boxplot(x='Churn', y=var, data=datos_expandido, palette="Set2", hue='Churn')
    plt.title(f'{var} vs Churn')
    plt.xlabel('Churn (0=No, 1=S√≠)')
    plt.ylabel(var)

plt.tight_layout()
plt.show()
````

##üîπ Conclusiones e Insights:

Segun lo estudiado podemos sacar algunos puntos relevantes, segun graficas y visualizaciones:

**Tasa de Churn alta:** Un 26.5% de los clientes han cancelado el servicio, lo que significa un numero de clientes muy considerable.

**Influencia del tipo de contrato:** Los contratos mensuales tienen la mayor√≠a de los casos de cancelacion, mientras que los contratos de uno o dos a√±os muestran tasas de abandono mucho menores. Esto sugiere que la permanencia de contrato m√°s larga est√° asociada con mayor retenci√≥n.

**Gastos totales y tiempo de permanencia:** Clientes con menos meses de contrato y menores gastos totales tienden a abandonar con mayor frecuencia. Esto podr√≠a indicar que los clientes nuevos o de bajo consumo no llegan a generar un v√≠nculo fiel con la empresa.

**M√©todo de pago y churn:** El m√©todo de pago Electronic Check tiene una proporci√≥n elevada de clientes que abandonan en comparaci√≥n con otros m√©todos como transferencias o tarjetas autom√°ticas. Esto puede estar asociado con la facilidad de cancelar o menor compromiso de pago autom√°tico.

**G√©nero sin diferencia significativa:** No se observan diferencias relevantes de churn entre hombres y mujeres, por lo que no parece ser un factor clave para la evasi√≥n.


Segun estos hallazgos se pueden proponer diferentes conceptos para manejar mejor las tazas de Churn:

**Fomentar contratos de mayor duraci√≥n**, ofreciendo incentivos o descuentos para migrar de mensual a anual o bianual.

**Programas de fidelizaci√≥n temprana para clientes nuevos**, ya que el mayor riesgo de abandono ocurre en los primeros meses.

**Revisar las condiciones del pago por Electronic Check**, ofreciendo beneficios por pasarse a m√©todos autom√°ticos.

**Campa√±as personalizadas para clientes con bajo gasto total**, buscando incrementar su uso y mejorar la percepci√≥n del servicio.
